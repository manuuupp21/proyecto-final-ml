```python
import pandas as pd
```

```python
# Paths
PATH_DIRECTORIO_DATOS = "../../data/raw"
PATH_DATASET_RESERVAS = f"{PATH_DIRECTORIO_DATOS}/dataset_practica_final.csv"
```

```python
# Cargar el dataset
df = pd.read_csv(PATH_DATASET_RESERVAS)
```

```python
#La columna arrival_date_month es la unica que no está en formato numérico de las fechas arrival_date_year y arrival_date_day_of_month
#Convertimos arrival_date_month a formato numérico
df['arrival_date_month'] = df['arrival_date_month'].replace(
{'January': 1,
'February': 2,
'March': 3,
'April': 4,
'May': 5,
'June': 6,
'July': 7,
'August': 8,
'September': 9,
'October': 10,
'November': 11,
'December': 12,
})
```

```python
#Se observa que las filas con valores nulos en la columna 'children' debe ser una errata o que se ha dejado vacío y decidimos cambiar esos valores por 0
df['children'] = df['children'].fillna(0)
```

```python
#Para country se decidie que el valor NA será representado por 'Unknown'
df['country'] = df['country'].fillna('Unknown')
```

```python
#Como la mayoria estan en variables NaN se decide crear una nueva variable que sea has_company indicando si tiene o no compañia y eliminar la columna company
df['has_company'] = df['company'].notna().astype(int)
df = df.drop(columns=['company'])
```

```python
#Con la columna agent no tengo todavia claro que hacer
df['agent'].value_counts(dropna=False)
df['agent'] = df['agent'].fillna('Unknown')
```

```python
#Existen columnas que tienen informacion del estado final de la reserva como
#reservation_status, reservation_status_date
#Estas columnas no se van a tener en cuenta para el modelo
df = df.drop(columns=['reservation_status', 'reservation_status_date'])
```

```python
#Guardar dataset en /processed
PATH_DIRECTORIO_DATOS_PROCESADOS = "../../data/processed"
PATH_DATASET_RESERVAS_PROCESADO = f"{PATH_DIRECTORIO_DATOS_PROCESADOS}/data_processed.csv"
df.to_csv(PATH_DATASET_RESERVAS_PROCESADO, index=False)
```
